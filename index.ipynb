```python
import pandas as pd

linkGit=    "https://github.com/Halabbi/PC2_BOTTONI/raw/refs/heads/main/HDL.csv"   #paste link here!

HDI=pd.read_csv(linkGit)
```

```python
HDI
```

```python
#eliminé las columnas que aparecían NaN innecesarios
columns_to_drop = ['Unnamed: 3', 'Unnamed: 5', 'Unnamed: 7', 'Unnamed: 9', 'Unnamed: 11', 'Unnamed: 13']
HDI_cleaned = HDI.drop(columns=columns_to_drop)
HDI_cleaned
```

```python
HDI_cleaned=HDI_cleaned.dropna(subset=['Unnamed: 2','Unnamed: 4', 'Unnamed: 6','Unnamed: 8','Unnamed: 10'])
HDI_cleaned
```

```python
#Cambién el nombre a los Unnamed
change={'Unnamed: 2':'','Unnamed: 4':'SDG3', 'Unnamed: 6':'SDG4.3','Unnamed: 8':'SDG4.4','Unnamed: 10':'SDG8.5','Unnamed: 12':'','Unnamed: 14':''}
HDI_cleaned.rename(columns=change, inplace=True)
HDI_cleaned
```

```python
#Se cambia los NaN por espacios en blanco sin alterar la tabla 
HDI_cleaned=HDI_cleaned.fillna('')
HDI_cleaned
```

```python
#Se reemplaza todas las comas por puntos
HDI_new= HDI_cleaned.replace(',', '.', regex=True)
HDI_new
```

```python
#Se agrega el dato que falta en la tabla
HDI_new.iloc[2, 1] = 'VERY HIGH HUMAN DEVELOPMENT'
HDI_new
```

```python
#Conteo de la fila Back
HDI_new['Back'] = range(-2, len(HDI_new) -2)

HDI_new
```

```python
#Se elimina los números innecesarios de las filas 
HDI_new.iloc[1, 0] = ""
HDI_new.iloc[0, 0] = ""
HDI_new.iloc[2, 0] = ""

HDI_new
```
